# 히스토그램 스트레칭

영상의 밝기값 범위를 확장(축소) 시킴으로써 영상의 대비를 향상시키는 방법

영상 픽셀들에 대해 선형확장함수를 적용해 구현함

new pixel = (old pixel - low)/(high - low) * 255



->End-in 탐색(상하위 절단)

일정한 양의 화소를 흰색이나 검정색으로 지정해 히스토그램의 분포를 좀 더 균일하게 만듦

*두개의 임계값(low high)사용*

dst = cv2.normalize(src, dst, alpha, beta, type_flag)

src: 정규화 이전 데이터

dst: 정규화 이후 데이터

alpha: 정규화할 하한값

beta: 정규화할 상한값

type_flag: 정규화 알고리즘 선택 플래그 상수(스트레칭의 경우 cv2.NORM_MINMAX로 설정)

```python
import cv2
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('/content/drive/MyDrive/교재실습영상/Copy of pentagon.bmp',0)
dst = cv2.normalize(src, None , 0,255, cv2.NORM_MINMAX)

src_hist = cv2.calcHist([src], [0], None, [256], [0,255])
plt.plot(src_hist, label="original")
dst_hist = cv2.calcHist([dst], [0], None, [256], [0,255])
plt.plot(dst_hist, label="normalized")
plt.legend() #범례표시
plt.show()
plt.subplot(1,2,1)
plt.imshow(src,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(dst,cmap='gray')
plt.show()
```
