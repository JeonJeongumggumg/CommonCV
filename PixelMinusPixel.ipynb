```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
image1 = cv2.imread('/content/drive/MyDrive/교재실습영상/Copy of aero2.bmp')
image2 = cv2.imread('/content/drive/MyDrive/교재실습영상/Copy of milkdrop.bmp')# 두사진의 크기 잘볼 것.
image_sub = cv2.subtract(image1,image2)
image_sub2 = cv2.absdiff(image1,image2)

cv2_imshow(image_sub) #결과차이를 유심히
cv2_imshow(image_sub2)
```

```python
#응용해서 차이점 발견하기
image3 = cv2.imread('/content/drive/MyDrive/교재실습영상/Copy of milkdrop1.bmp')
image4 = cv2.imread('/content/drive/MyDrive/교재실습영상/Copy of milkdrop.bmp')

image_sub3 = cv2.absdiff(image3,image4)

cv2_imshow(image3)
cv2_imshow(image4)

cv2_imshow(image_sub3) #결과차이를 유심히
```
