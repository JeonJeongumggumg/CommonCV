# 임계값적용(밝기, 색상차이를 통해 전경과 배경을 분리하는 기본방법)

사용자에 의해 임계값이 결정 혹은 영상으로부터의 히스토그램을 생성해 자동결정

전역 임계값 적용(Otsu)

각 클래스의 분산을 최소화하는 임계값을 최적의 임계값으로 사용
(모든 이미지에 최적 임계값을 지정x)
(히스토그램이 쌍봉우리 형태로 나타날때 가장 효과적인듯)

ret, out = cv2.threshold(img,*threshold*,value,type)

ex) ret,dst = cv2.threshold(src,120,255,cv2.THRESH_BINARY)

Two value of return(ret:임계값, out:결과영상)

이진화 TYPE

THRESH_BINARY

THRESH_BINARY_INV

THRESH_TRUNC(임계보다 크면 기존값으로 유지)

THRESH_TOZERO(임계보다 크면 기존값으로 유지, 작으면 0으로)

THRESH_TOZERO_INV(임계값보다 크면 0으로, 작으면 기존값 유지)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#이미지를 그레이 스케일로 읽기
img = cv2.imread('/content/drive/MyDrive/교재실습영상/Copy of camera.bmp',0)
#경계값을 지정
th1, t_130 = cv2.threshold(img,130,255,cv2.THRESH_BINARY)
#경계값 지정x OTSU알고리즘 선택
th2, t_otsu = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print('otsu threshold', th2)

plt.subplot(121)
plt.imshow(t_130, cmap='gray')
plt.subplot(122)
plt.imshow(t_otsu, cmap='gray')
```
