# 픽셀 더하기 (OpenCV 함수를 이용)

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
image1 = cv2.imread('/content/drive/MyDrive/교재실습영상/Copy of aero2.bmp')
number = np.ones_like(image1) * 100 #동일크기의 벡터를 생성해 1로 초기화 후 반환

image_add = cv2.add(image1, number)
image_sub = cv2.subtract(image1, number)

cv2_imshow(image_add)
cv2_imshow(image_sub)
```
