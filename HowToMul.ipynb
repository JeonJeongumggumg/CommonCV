픽셀 곱셈 연산하기

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
src = cv2.imread('/content/drive/MyDrive/교재실습영상/Copy of lena.jpg',0)
number2 = np.ones_like(src)*2
mul = cv2.multiply(src,number2)
cv2_imshow(mul)
```

--> 원본 사이즈가 너무 커서 결과가 너무 크게 출력

# matplotlib로 영상 사이즈를 조절하기

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3,3)) #사이즈 설정
plt.imshow(src,cmap='gray') #영상을 출력
```
