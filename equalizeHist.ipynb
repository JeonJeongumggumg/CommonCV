# 히스토그램 평활화(Histogram Equalization)

어두컴컴한 영상의 히스토그램을 조절해 명암분포를 균일하게 만듦

영상의 밝기 분포를 재분배해 명암대비를 최대화

영상의 검출 특성을 증가시킴

--> 대비가 커질 수록 인간의 인지도가 증가

히스토 생성 - 정규화 히스토로 변형 - 변형한 히스토로 입력영상을 매핑



cv2.equalizeHist(src)

src: grayscale data

```python
import cv2
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('/content/drive/MyDrive/교재실습영상/Copy of pentagon.bmp',0)

dst = cv2.equalizeHist(src)

src_Hist = cv2.calcHist([src],[0],None,[256],[0,256])
plt.plot(src_Hist,label="original")
dst_Hist = cv2.calcHist([dst],[0],None,[256],[0,256])
plt.plot(dst_Hist,label="eq")
plt.legend() #범례표시
plt.show()
plt.subplot(1,2,1)
plt.imshow(src,cmap = 'gray')
plt.subplot(1,2,2)
plt.imshow(dst,cmap = 'gray')
plt.show
```
