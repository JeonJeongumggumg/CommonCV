```python
import cv2
from google.colab.patches import cv2_imshow
path1 = '/content/drive/MyDrive/교재실습영상/Copy of aero2.bmp'
path2 = '/content/drive/MyDrive/교재실습영상/Copy of milkdrop.bmp'
```

```python
img1 = cv2.imread(path1)
img2 = cv2.imread(path2)
result = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
cv2_imshow(result)
```

```python
img3 = cv2.imread(path1)
img4 = cv2.imread(path2)
result1 = cv2.addWeighted(img3, 0.2, img4, 0.8, 0)
cv2_imshow(result1)
```
